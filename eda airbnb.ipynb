# **Project Summary -**


The rapid growth of online rental platforms like Airbnb has created massive volumes of data, offering valuable opportunities for hosts and platform managers to understand customer behavior, listing trends, and revenue drivers. The primary objective of this project was to perform Exploratory Data Analysis (EDA) on an Airbnb dataset to extract meaningful insights that could inform business decisions, optimize listing performance, and improve the user experience on the platform.

Data Overview
The dataset used consists of thousands of Airbnb listings, containing variables such as price, room_type, location, number_of_reviews, reviews_per_month, availability_365, and more. An initial inspection revealed:

Over 48,000 listings across different room types and neighborhoods.

Data types ranging from categorical (e.g., room_type, neighbourhood) to numerical (e.g., price, reviews_per_month).

Some missing values and duplicate entries, which were cleaned during the data wrangling process.

Data Cleaning and Preparation
Before conducting in-depth analysis, the dataset was cleaned to ensure accuracy and usability:

Removed duplicate records.

Handled missing values by filling or dropping, depending on the column.

Converted price-related columns into numeric format and removed outliers to ensure visualizations were meaningful.

Selected key variables for focused analysis: price, number_of_reviews, reviews_per_month, availability_365, room_type, and neighbourhood_group.

Key Insights from Analysis:
1. Pricing Trends
Most listings are priced between $50–$150, and this range consistently shows the highest number of reviews.

Luxury-priced listings are fewer and often have low engagement, suggesting a narrow target market or pricing inefficiencies.

2. Room Type Performance
Private rooms dominate in terms of frequency and guest engagement.

Entire homes/apartments are popular but show a wider variance in availability and price.

Shared rooms have minimal presence and limited reviews, indicating lower demand.

3. Availability and Booking
Listings with higher availability_365 tend to receive more reviews, showing a positive relationship between availability and booking activity.

Encouraging hosts to make their listings available more often can directly increase visibility and revenue.

4. Guest Review Behavior
number_of_reviews and reviews_per_month are strongly correlated.

Neighborhoods with consistently high reviews should be prioritized in platform marketing strategies.

5. Geographic and Seasonal Insights
Certain neighborhoods are more saturated, while others present untapped opportunities for growth.

Listings in top-performing areas receive the most guest activity and could benefit from featured placement.

Visualizations Used
To uncover and communicate insights, multiple visualizations were generated:

Bar plots for distribution of listings by room type and location.

Box plots to explore price variability across room types.

Heatmaps to show correlation among numeric variables.

Pair plots to detect variable relationships and outliers.

Scatter plots and line charts to understand trends in reviews, availability, and pricing.

Business Recommendations
Based on the insights, the following actions are suggested to Airbnb or its hosts:

Optimize pricing strategies using historical booking data.

Encourage hosts to increase listing availability.

Promote private rooms for consistent engagement.

Educate hosts on the importance of reviews and listing completeness.

Use neighborhood-level performance to guide platform promotions and featured listings.

Conclusion
The EDA provided valuable, data-driven insights that can be used to improve business performance for both Airbnb and its hosts. It highlights how pricing, availability, location, and listing type all impact customer engagement. By aligning strategies with these insights, Airbnb can drive greater customer satisfaction, increase booking rates, and expand its market presence in a sustainable way.

````python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/eda data project/Copy of Airbnb NYC 2019.csv') # update the path if needed

# -----------------------------------
# 🔹 1. Remove Duplicates
# -----------------------------------
df = df.drop_duplicates()
print(f"After removing duplicates: {df.shape}")

# -----------------------------------
# 🔹 2. Handle Missing Values
# -----------------------------------
# Check missing values
missing = df.isnull().sum()
print("Missing Values:\n", missing[missing > 0])

# Fill 'reviews_per_month' with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Fill 'last_review' with a placeholder or drop (optional)
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Drop rows where 'name' or 'host_name' is missing (optional)
df = df.dropna(subset=['name', 'host_name'])

# -----------------------------------
# 🔹 3. Convert Data Types
# -----------------------------------
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')  # already done above

# -----------------------------------
# 🔹 4. Fix Invalid Values
# -----------------------------------
# Remove listings with 0 or negative price
df = df[df['price'] > 0]

# Cap outliers in price (optional)
df = df[df['price'] < 1000]  # remove extreme luxury listings

# Optional: Remove extreme minimum_nights
df = df[df['minimum_nights'] < 365]

# -----------------------------------
# 🔹 5. Reset Index After Cleaning
# -----------------------------------
df = df.reset_index(drop=True)

# -----------------------------------
# ✅ Final Cleaned Data Overview
# -----------------------------------
print("Cleaned dataset shape:", df.shape)
print("\nCleaned dataset preview:\n", df.head())
````

````python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/eda data project/Copy of Airbnb NYC 2019.csv')

# -----------------------------------
# 🔹 1. Remove Duplicates
# -----------------------------------
df = df.drop_duplicates()
print(f"After removing duplicates: {df.shape}")

# -----------------------------------
# 🔹 2. Handle Missing Values
# -----------------------------------
# Check missing values
missing = df.isnull().sum()
print("Missing Values:\n", missing[missing > 0])

# Fill 'reviews_per_month' with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Fill 'last_review' with a placeholder or drop (optional)
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Drop rows where 'name' or 'host_name' is missing (optional)
df = df.dropna(subset=['name', 'host_name'])

# -----------------------------------
# 🔹 3. Convert Data Types
# -----------------------------------
# No explicit data type conversions needed here as pd.to_datetime is used above

# -----------------------------------
# 🔹 4. Fix Invalid Values
# -----------------------------------
# Remove listings with 0 or negative price
df = df[df['price'] > 0]

# Cap outliers in price (optional)
df = df[df['price'] < 1000]  # remove extreme luxury listings

# Optional: Remove extreme minimum_nights
df = df[df['minimum_nights'] < 365]

# -----------------------------------
# 🔹 5. Reset Index After Cleaning
# -----------------------------------
df = df.reset_index(drop=True)

# -----------------------------------
# ✅ Final Cleaned Data Overview
# -----------------------------------
print("Cleaned dataset shape:", df.shape)
print("\nCleaned dataset preview:\n", df.head())
````
